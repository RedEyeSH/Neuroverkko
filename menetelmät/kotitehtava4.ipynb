# 4. Text data and preprocessing

## Word Embeddings

Word embeddings represent words as dense numerical vectors. These vectors capture semantic meaning, allowing similar words to appear close to each other in the vector space. This makes it possible to analyze relationships between words using mathematical operations.

---

## Importing Libraries

```python
import numpy as np
```

---

## Loading Pretrained GloVe Embeddings

The pretrained GloVe embeddings are loaded from a text file where each line contains a word followed by its vector representation. In this case, each word is represented as a 100-dimensional vector.

```python
embeddings = {}

path = "./glove.2024.wikigiga.100d/wiki_giga_2024_100_MFT20_vectors_seed_2024_alpha_0.75_eta_0.05.050_combined.txt"

with open(path, encoding="utf-8") as f:
    for line in f:
        values = line.split()
        
        try:
            word = values[0]
            vector = np.asarray(values[1:], dtype="float32")
            
            if len(vector) == 100:
                embeddings[word] = vector
                
        except:
            continue

print("Words count:", len(embeddings))

words = list(embeddings.keys())
matrix = np.array(list(embeddings.values()))
```

---

## Vector Arithmetic

Word embeddings make it possible to do simple calculations with words and explore the relationships between them. For example, the expression **woman − man + king** can be understood by first removing the idea of “man” from “woman” and then adding the idea of “king”. The result is typically **queen**, which shows how the model captures relationships like “man is to woman as king is to queen”.

---

```python
vector = embeddings["woman"] - embeddings["man"] + embeddings["king"]
```

---

## Finding Similar Words

To interpret the result, cosine similarity is used to find the nearest words to the computed vector. This measures how similar two vectors are in direction.

```python
from sklearn.metrics.pairwise import cosine_similarity 

def find_nearest(vec, words, matrix, top_n=5, exclude=None):
    if exclude is None:
        exclude = []

    similarities = cosine_similarity([vec], matrix)[0]
    top_indices = np.argsort(similarities)[::-1]
    
    result = []
    for i in top_indices:
        word = words[i]
        if word not in exclude:
            result.append(word)
        if len(result) == top_n:
            break
            
    return result
```

---

## Main Result

The nearest word to the computed vector is found. The original words are excluded to avoid trivial matches.

```python
vector = embeddings["woman"] - embeddings["man"] + embeddings["king"]

result = find_nearest(vector, words, matrix, exclude=["man", "woman", "king"])

print("Top results for woman - man + king:")
print(result[0])
```

The result is typically "queen", which shows that the model captures the relationship between gender and royalty.

---

## Additional Examples

The same idea can be tested with other word relationships.

```python
print("Paris - France + Italy -->", find_nearest(
    embeddings["paris"] - embeddings["france"] + embeddings["italy"],
    words, matrix,
    exclude=["paris", "france", "italy"]
)[0], "\n")

print("Walking - Walk + Run -->", find_nearest(
    embeddings["walking"] - embeddings["walk"] + embeddings["run"],
    words, matrix,
    exclude=["walking", "walk", "run"]
)[0], "\n")

print("Big - Bigger + Smaller -->", find_nearest(
    embeddings["big"] - embeddings["bigger"] + embeddings["smaller"], 
    words, matrix, 
    exclude=["big", "bigger", "smaller"]
)[0], "\n")
```

---

## Results and Discussion

The main analogy produces the expected result (“queen”), which shows that word embeddings can capture meaningful relationships between words. The additional examples suggest that the model also learns grammatical and contextual patterns, although the results are not always equally accurate.

---

## Conclusion

Word embeddings are a way to represent words as numbers so that their meanings can be compared and analysed. By placing words into a shared vector space, it becomes possible to explore relationships between them using simple mathematical operations. Although the method is not perfect, it captures useful patterns in language and is widely used in natural language processing.
