## Word Embeddings

Word embeddings represent words as dense numerical vectors. These vectors capture semantic meaning, allowing similar words to appear close to each other in the vector space. This makes it possible to analyze relationships between words using mathematical operations.

---

## Importing Libraries

NumPy is used for numerical operations and handling vector data.

```python
import numpy as np
```

---

## Loading Pretrained GloVe Embeddings

The pretrained GloVe embeddings are loaded from a file. Each line contains a word followed by its vector representation. In this case, each word is represented by a 100-dimensional vector.

```python
embeddings = {}

path = "./glove.2024.wikigiga.100d/wiki_giga_2024_100_MFT20_vectors_seed_2024_alpha_0.75_eta_0.05.050_combined.txt"

with open(path, encoding="utf-8") as f:
    for line in f:
        values = line.split()
        
        try:
            word = values[0]
            vector = np.asarray(values[1:], dtype="float32")
            
            if len(vector) == 100:
                embeddings[word] = vector
                
        except:
            continue

print("Words count:", len(embeddings))
```

---

## Vector Arithmetic

Word embeddings allow simple mathematical operations to capture relationships between words. The expression:

**woman** - **man** + **king**

is expected to represent the female equivalent of a king.

```python
vector = embeddings["woman"] - embeddings["man"] + embeddings["king"]
```

---

## Finding Similar Words

To interpret the result, cosine similarity is used to find the closest words to the computed vector. This measures how similar two vectors are in direction.

```python
from sklearn.metrics.pairwise import cosine_similarity 

def find_nearest(vec, embeddings, top_n=5, exclude=[]):
    words = list(embeddings.keys())
    matrix = np.array(list(embeddings.values()))
    
    similarities = cosine_similarity([vec], matrix)[0]
    top_indices = np.argsort(similarities)[::-1]
    
    result = []
    for i in top_indices:
        word = words[i]
        if word not in exclude:
            result.append(word)
        if len(result) == top_n:
            break
            
    return result[0]
```

---

## Main Result

The nearest word to the computed vector is found. The original words are excluded to avoid trivial matches.

```python
vector = embeddings["woman"] - embeddings["man"] + embeddings["king"]

result = find_nearest(vector, embeddings, exclude=["man", "woman", "king"])

print("Result for woman - man + king:")
print(result)
```

The result is typically "queen", which shows that the model captures the relationship between gender and royalty.

---

## Additional Examples

The same idea can be tested with other word relationships.

```python
print("Paris - France + Italy")
print(find_nearest(embeddings["paris"] - embeddings["france"] + embeddings["italy"], embeddings) + "\n")

print("Walking - walk + run")
print(find_nearest(embeddings["walking"] - embeddings["walk"] + embeddings["run"], embeddings) + "\n")

print("Big - bigger + smaller")
print(find_nearest(embeddings["big"] - embeddings["bigger"] + embeddings["smaller"], embeddings) + "\n")
```

---

## Results and Discussion

The main analogy produces the expected result ("queen"), showing that word embeddings can capture meaningful semantic relationships.
The additional examples demonstrate that embeddings also encode grammatical and contextual patterns, although with varying accuracy.

---

## Conclusion

Word embeddings provide a powerful way to represent language numerically. By mapping words into a vector space, they allow relationships between words to be explored using simple mathematical operations. While not perfect, they reveal meaningful structures in language and are widely used in natural language processing tasks.
