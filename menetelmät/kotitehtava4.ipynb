# 4. Text data and preprocessing

Word embeddings represent words as numerical vectors. These vectors capture semantic meaning, so that similar words are located close to each other in the vector space.

To work with vectors, NumPy is used for efficient numerical operations.

```python
import numpy as np
```

The pretrained GloVe embeddings are loaded from a file. Each line in the file contains a word followed by its vector representation. The vectors used here have 100 dimensions.

```python
embeddings = {}

with open("./glove.2024.wikigiga.100d/wiki_giga_2024_100_MFT20_vectors_seed_2024_alpha_0.75_eta_0.05.050_combined.txt", encoding="utf-8") as f:
    for line in f:
        values = line.split()
        
        try:
            word = values[0]
            vector = np.asarray(values[1:], dtype="float32")
            
            if len(vector) == 100:
                embeddings[word] = vector
                
        except:
            continue

print("Words count:", len(embeddings))
```

The expression **woman** - **man** + **king** is expected to represent the concept of a female king.

```python
vector = embeddings["woman"] - embeddings["man"] + embeddings["king"]
```

## Find nearest words using cosine similarity

To interpret the result, the closest word to the computed vector is searched. Cosine similarity is used to measure how similar two vectors are.

```python
from sklearn.metrics.pairwise import cosine_similarity 

def find_nearest(vec, embeddings, top_n=5, exclude=[]):
    words = list(embeddings.keys())
    matrix = np.array(list(embeddings.values()))
    
    similarities = cosine_similarity([vec], matrix)[0]
    top_indices = np.argsort(similarities)[::-1]
    
    result = []
    for i in top_indices:
        word = words[i]
        if word not in exclude:
            result.append(word)
        if len(result) == top_n:
            break
            
    return result[0]
```

The nearest word to the computed vector is then found. The original words are excluded to avoid trivial matches.

```python
vector = embeddings["woman"] - embeddings["man"] + embeddings["king"]

result = find_nearest(vector, embeddings, exclude=["man", "woman", "king"])

print("Result for woman - man + king:")
print(result)
```

The same idea can be tested with other word relationships to see how well the embeddings capture different types of meaning.

```python
print("Rome - France + Italy")
print(find_nearest(embeddings["paris"] - embeddings["france"] + embeddings["italy"], embeddings) + "\n")

print("Walking - walk + run")
print(find_nearest(embeddings["walking"] - embeddings["walk"] + embeddings["run"], embeddings) + "\n")

print("Big - bigger + smaller")
print(find_nearest(embeddings["big"] - embeddings["bigger"] + embeddings["smaller"], embeddings) + "\n")
```
