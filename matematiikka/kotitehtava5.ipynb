## Tehtävä 1

Tässä piirretään funktion p(x) kuvaaja, jotta nähdään sen muoto ja mahdolliset minimit.

```python
def p(x):
    y = x**4 + 5 * x**3 - 7 * x
    return y

xx = np.linspace(-5, 2, 1000)
f = p(xx)
plt.plot(xx, f)
plt.grid()
plt.show()
```

Lasketaan seuraavassa osassa funktion derivaatta ja käytetään **gradient descent** minimin etsimiseen

```python
# x:n derivaatta
def dp(x):
    y = 4 * x**3 + 15 * x**2 - 7
    return y

def gradient_descent(x_start, alpha, iterations=1000):
    x = x_start
    
    for _ in range(iterations):
        gradient = dp(x)
        x = x - alpha * gradient
            
    return x

alpha = 0.01
minimi1 = gradient_descent(x_start=0, alpha=alpha)
minimi2 = gradient_descent(x_start=-4, alpha=alpha)

print(f"Vastaus: x = {minimi2:.6f} tai x = {minimi1:.6f}")

x = 1
path_x = np.array([x])
```

Piirretään gradient descent ja funktion kuvaaja.

```python
path_y = p(path_x)
plt.plot(path_x, path_y, '.')

xx = np.linspace(-4, 2, 1000)
f = p(xx)
plt.plot(xx, f)
plt.grid()
plt.show()
```

## Tehtävä 2

Tässä toteutetaan gradient descent, jossa etsitään funktion f(x, y) minimiä päivittämällä x- ja y-arvoja osittaisderivaattojen avulla ja tallentamalla polku.

```python
# alkuperäinen funktio
def f(x, y):
    return 3*x**2 + 3*x*y + 2*y**2 + 3*x - 5*y

# osittaisderivaatat
def df(x, y):
    dx = 6*x + 3*y + 3 # x:n derivaatta
    dy = 3*x + 4*y - 5 # y:n derivaatta
    return dx, dy

def gradient_descent(x_start, alpha, iterations=1000):
    x = x_start
    
    for _ in range(iterations):
        gradient = dp(x)
        x = x - alpha * gradient
            
    return x

x, y = 0, 0
alpha = 0.05
path_x = np.array([x])
path_y = np.array([y])

for i in range(200):
    gx, gy = df(x, y)
    x = x - alpha * gx
    y = y - alpha * gy
    path_x = np.append(path_x, x)
    path_y = np.append(path_y, y)
    

print(f"Vastaus: x = {x:.1f},  y = {y:.1f}")
```

Tässä piirretään funktion f(x, y) kontuurikuva ja päälle gradient descent, jotta nähdään miten piste etenee kohti minimiä.

```python
xx = np.linspace(-7, 7, 100)
yy = np.linspace(-5, 5, 100)
(X, Y) = np.meshgrid(xx, yy)

Z = 3*X**2 + 3*X*Y + 2*Y**2 + 3*X - 5*Y

plt.contour(X, Y, Z)
plt.plot(path_x, path_y, 'r+')
plt.axis('equal')
plt.show()
```
